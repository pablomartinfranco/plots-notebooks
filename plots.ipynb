```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
# %matplotlib widget
# %matplotlib inline
# %matplotlib notebook
plt.rcParams['figure.dpi'] = 120
```

```python
# Normal distribution with mean=0 and standard deviation=1

mu = 0
std = 1
x = np.linspace(start=-4, stop=4, num=100)
y = stats.norm.pdf(x, mu, std)
plt.plot(x, y)
plt.show()
```

```python
# Adding Grid Lines

fig = plt.figure()
ax = plt.axes()
plt.grid()
plt.show()
```

```python
# The color

fig = plt.figure()
ax = plt.axes()
plt.grid(visible=True, color='r')
```

```python
# Adding Grid Lines

fig = plt.figure()
ax = plt.axes()
plt.grid(visible=True, color='aqua', alpha=0.3, linestyle='-.', linewidth=2)
```

```python
# The color and linestyle arguments

fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 5, 100)
plt.plot(x, np.sin(x), color='Indigo', linestyle='--', linewidth=3)
plt.grid(visible=True, color='aqua', alpha=0.3, linestyle='-.', linewidth=2)
plt.show()
```

```python
# This combining of color and linestyle will not work with the grid

fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 5, 100)
plt.plot(x, np.sin(x), 'c--', linewidth=3)
plt.grid(visible=True, color='aqua', alpha=0.3, linestyle='-.', linewidth=2)
plt.show()
```

```python
# plt.minorticks_on() displays minor ticks on the axes. The disadvantage is that it may reduce performance.

x = np.linspace(0, 7, 100)
line1, = plt.plot(x, np.sin(x), label='sin')
line2, = plt.plot(x, np.cos(x), label='cos')
plt.legend(handles=[line1, line2], loc='lower right')
#major grid lines
plt.grid(visible=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)
#minor grid lines
plt.minorticks_on()
plt.grid(visible=True, which='minor', color='beige', alpha=0.8, ls='-', lw=1)
plt.xlabel('x', fontsize=15)
plt.ylabel('PDF', fontsize=15)
plt.show()
```

```python
# With Axes Limits

fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 5, 100)
plt.plot(x, np.sin(x), 'c--', linewidth=3)
plt.xlim(-1, 6)
plt.ylim(-1.25, 1.25)
plt.grid(visible=True, color='aqua', alpha=0.3, linestyle='-.', linewidth=2)
plt.show()
```

```python
# Tick Parameters

fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 5, 100)
plt.plot(x, np.sin(x), 'c--', linewidth=3)
plt.xlim(-1, 6)
plt.ylim(-1.25, 1.25)
plt.tick_params(axis='both', direction='out', length=6, width=2, labelcolor='b', colors='r', grid_color='gray', grid_alpha=0.5)
plt.grid()
plt.show()
```

```python
# Adding Legend

x = np.linspace(0, 7, 100)
line1, = plt.plot(x, np.sin(x), label='sin')
line2, = plt.plot(x, np.cos(x), label='cos')
plt.legend(handles=[line1, line2], loc='best')
plt.grid(visible=True, color='aqua', alpha=0.6, linestyle='dashdot')
plt.show()
```

```python
mu = 0
std = 1
x = np.linspace(start=-4, stop=4, num=100)
y = stats.norm.pdf(x, mu, std)
plt.plot(x, y, label='PDF')
plt.xlabel('x', fontsize=15)
plt.ylabel('PDF', fontsize=15)
plt.grid(visible=True, which='major', color='DarkTurquoise', alpha=0.4, linestyle=':', linewidth=2)
plt.minorticks_on()
plt.grid(visible=True, which='minor', color='beige', alpha=0.2, linestyle='-', linewidth=2)
plt.legend()
plt.savefig('Normal Distribution.png')
plt.show()
```

```python
# Conversion to Centimeters

mu = 0
std = 1
x = np.linspace(start=-4, stop=4, num=100)
y = stats.norm.pdf(x, mu, std)
plt.plot(x, y, label='PDF')
plt.xlabel('x', fontsize=15)
plt.ylabel('PDF', fontsize=15)
plt.grid(visible=True, color='DarkTurquoise', alpha=0.2, linestyle=':', linewidth=2)
plt.rcParams['figure.figsize'] = [10/2.54, 8/2.54]
plt.legend()
plt.show()
```
