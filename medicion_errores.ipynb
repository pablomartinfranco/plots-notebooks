```python
import matplotlib.pyplot as plt

# Data for the intervals and errors
intervals = ['Probeta', 'Regla', 'Calibre']
values = [5000, 5300, 4990]
errors = [2000, 900, 50]

# Define the x-axis values
x = range(len(intervals))

# Set the figure size
plt.figure(figsize=(5, 10))

# Plot the bar chart with adjusted width
plt.bar(x, values, yerr=errors, tick_label=intervals, color='skyblue', edgecolor='black', width=0.5)

# Set the y-axis limits and ticks
plt.ylim(2900, 7100)
plt.yticks(range(3000, 7001, 200))

# Add horizontal rulers at a resolution of 100 mm^3
plt.grid(axis='y', linestyle='--')

# Add horizontal lines to the tolerance edges
for i, val in enumerate(values):
    lower_edge = val - errors[i]
    upper_edge = val + errors[i]
    plt.axhline(y=lower_edge, color='red', linestyle='--')
    plt.axhline(y=upper_edge, color='red', linestyle='--')

# Add labels and title
plt.xlabel('Intervalos de error')
plt.ylabel('Volumen (mm^3)')
plt.title('Comparacin de intervalos de error')

# Display the chart
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data for the intervals and errors
intervals = ['Probeta', 'Regla', 'Calibre']
values = [5000, 5300, 4990]
errors = [2000, 900, 50]

# Define the x-axis values
x = range(len(intervals))

# Set the figure size
plt.figure(figsize=(5, 10))

# Plot the tolerance intervals and connect the edges with a line
for i in range(len(intervals)):
    # Calculate the left and right edges of the tolerance interval
    left_edge = values[i] - errors[i]
    right_edge = values[i] + errors[i]

    # Plot the tolerance interval as a horizontal line
    # plt.hlines([left_edge, right_edge], x[i] - 0.25, x[i] + 0.25, colors='red', linewidth=2)

    # Add 5 mm horizontal lines to the tolerance edges and center
    plt.hlines([left_edge - 5, right_edge + 5, values[i]], x[i] - 0.25, x[i] + 0.25, colors='black', linewidth=1)

    # Connect the edges with a vertical line
    plt.vlines(x[i], left_edge, right_edge, colors='black', linewidth=2)

# Set the x-axis tick labels
plt.xticks(x, intervals)

# Set the y-axis limits and ticks
plt.ylim(2900, 7100)
plt.yticks(range(3000, 7001, 200))

# Add horizontal rulers at a resolution of 100 mm^3
plt.grid(axis='y', linestyle='--')

# Add labels and title
plt.xlabel('Intervalos de error')
plt.ylabel('Volumen (mm^3)')
plt.title('Comparacin de intervalos de error')

# Display the chart
plt.show()
```
